## **`for` loops**
`for` loops are constructs that we can use to repeat code.

For example, let's print something out multiple times.

```python
# print five times
for i in range(10):
    print(f"{i + 1}. Mr. Ubial is sooo cool.")
```

```python
for i in range(5):
    print(i)
    print("----")
```

```python
# Nested loop
for i in range(5):
    for j in range(5):
        print(f"{i} {j}")
        # goes through j 5 times before going to i again
```

```python
# print the numbers from 1 to 10
for i in range(10):
    print(i + 1)
```

```python
# print numbers from 1 to 10, v2
# range ()
# range(<start>, <stop>, <>)
# stop cuts off last

for i in range(1, 11):
    print(i)
```

### Counting by numbers other than one
There are ways we can use `range` to count by numbers other than one.

```python
# count even numbers 2 to 10

# range(<start>, <stop-exc>, <step>)
for i in range(2, 11, 2):
    print(i)
```

```python
# count even numbers 2 to 10, v2

for i in range(5):
    print((i + 1) * 2)
```

```python
# count backwards from 10 to 1

for i in range(10, 0, -1):
    print(i)
```

```python
list(range(0, 10, 2))
```

# **Lists and Loops**

## Lists

Lists are a type of **data**

* an ordered set of values

* use `[]` to create our list
* use`,` to separate individual elements

```python
# conventions
# ------
# name your lists as a PLURAL name

# calculate the average percentage

percentages = [60, 34, 86, 93, 51]

total = 0 # initializing a variable. doesn't have to be 0

# for <element> in <list>:
for percentage in percentages: # for each percentage in the percentages list
    total += percentage
    
print(f"The average is {total / len(percentages)}%")
# len is short for length i think!!!
```

```python
questions = [
    "What is the best food? ",
    "What's your favourite season? ",
    "What's the meaning of life? ",
]

for question in questions:
    print(question)
    input()
    print("Oooo. I see.")
```

### Application of `for` Loops
We want to add up all the numbers from 1 to 100.

```python
# loop from all numbers from 1 to 100
# take each number and add to a total
# print the total at the end

total = 0

for i in range(1, 101):
    total += i
    
print(total)
```

```python
sum(list(range(1, 101)))   # gives all values 1-10

# sum can be used on lists : )

# cool to make lines smaller! less effort, easier to work with.

# but there is more description with the for loop so it depends on what works best
```

## **`while` loops**
`while` loops are used when you don't know how many times you want to loop.

```python
# loop until something happens

# infinite loop

#      will repeat until not True
while True:
    print("Hi")
```

```python
# input validation

done = False

while not done:
    response = input("Do you want to quit? (y/n) ")
    if response.lower().strip(".?/!") in ["y", "yes"]:
        done = True
    
print("Thanks! We're done!")
```

### `random` module and modules in general
**Modules** are tools that others have created to make our lives easier.

For example, `random` includes a library of tools that have to do with randomness.

```python
import random

# Implement a coin flip

# random.random() -> [0.0, 1.0]

if random.random() < 0.5:
    print("Heads")
else:
    print("Tails")
```

```python
# generates a float between
# 0 and 1

random.random()

# 1, 100 FLOATS
random.uniform(1, 101)

# 1, 100 integers
random.randrange(1, 101)
```
